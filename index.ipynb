## Introduction

In the unit on infinite series you spend a lot of time learning different tests for convergence. One of the important tests is known as the Comparison Test. The purpose the comparison test is to allows you use to use known convergent and divegent series in order to determine whether a *new* series converges or diverges. 

For example, by the p-series test you know that the series
$$
\sum_{k=1}^\infty \frac1{k^2}
$$
converges. Can we use this fact to tell us anything about series that "look like" the known series? For example, can we say that
$$
\sum_{k=1}^\infty \frac1{k^2+10k} \quad\text{or}\quad \sum_{k=1}^\infty \frac1{k^2-0.5}
$$
converge? It turns out that we can immediately conclude that the first series converges using the known series and the Comparison Test. We don't know any information (yet) about the second series.

**Theorem (Comparison Test)**

*Let $\sum_{k=1}^\infty a_n$ and $\sum_{k=1}^\infty b_n$ be infinite series with **postitive** terms where it is known whether $\sum_{k=1}^\infty b_n$ converges or diverges.*

- *If $a_n \le b_n$ for all $n$ and $\sum_{k=1}^\infty b_n$ **converges**, then so does $\sum_{k=1}^\infty a_n$*
- *If $a_n \ge b_n$ for all $n$ and $\sum_{k=1}^\infty b_n$ **diverges**, then so does $\sum_{k=1}^\infty a_n$*

The purpose of this notebook is to help you visualize why the comparison test is true.
